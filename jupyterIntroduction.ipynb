build shareable documents that combine that 
* Code with

* Video

## Your first cell with python code

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
t = np.arange(0.0, 5.0, 0.01)
s = np.cos(4*np.pi*t)
plt.plot(t, s, linewidth=2.0, color='b')

plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('A title')
plt.grid(True)
plt.savefig("test.png")
plt.show()
```

## The jupyter notebook: A web-based UI for writing and running code

The IPython Notebook is an open source (BSD) tool for telling **stories** with code and data that are:

* Interactive
* Exploratory
* Collaborative
* Open
* Reproducible

### Original workflow

* slow (read whole data file each time, lots of context switching)
* version controlled analysis, but not commentary, difficult to 'go back to'
* Automating requires non-trivial additional dev

### Now workflow

 * Speedups primarily from no context switching, interactivity, and reusable data loading.
 * Reproducible, literate, annotatable, *auditable*.

## Extensions (in addition to kernels)

Not just Python code though! Though cell magics (`%%`) the jupyter notebooks supports running code in other languages:

* R
* ferret (read https://github.com/ipython/ipython/wiki/Extensions-Index#ferret)
* Octave
* Cython
* Bash
* Perl
* Ruby
* Julia
* latex
* etc.

You can enter latex directly with the %%latex cell magic:

```python
%%latex
\begin{aligned}
\nabla \times \vec{\mathbf{B}} -\, \frac1c\, \frac{\partial\vec{\mathbf{E}}}{\partial t} & = \frac{4\pi}{c}\vec{\mathbf{j}} \\
\nabla \cdot \vec{\mathbf{E}} & = 4 \pi \rho \\
\nabla \times \vec{\mathbf{E}}\, +\, \frac1c\, \frac{\partial\vec{\mathbf{B}}}{\partial t} & = \vec{\mathbf{0}} \\
\nabla \cdot \vec{\mathbf{B}} & = 0
\end{aligned}
```

## Essential Shortcuts

* ``Esc/Enter``: Mode Switch
* ``j/k``: Move up/down
* Execute Cells
    * ``Shift-Enter``: Run and go down
    * ``Alt-Enter``: Run and make new
    * ``Control-Enter``: Run in place
* ``a/b``: Insert cell above/below
* ``x``: cut cell
* Cell mode switch:
    * `r`: raw
    * `m`: markdown
    * `y`: p**y**thon code

## Building slides

* Turn on the 'slideshow' cell toolbar
* Types:
    * Slide: start a new slide
    * ``-``: Continue a slide
    * Sub-Slide: Make a 'down' slide
    * Fragment: Make a 'bullet' type incoming slide
    * Skip: keep in the notebook, not the deck
    * Notes: speaker notes
    
Then type

    jupyter-nbconvert --to slides myNotebook.ipynb

or by including javascript from reveal.js (html presentation framework)
    
    jupyter-nbconvert --to slides --reveal-prefix https://cdnjs.cloudflare.com/ajax/libs/reveal.js/3.4.1 myNotebook.ipynb
