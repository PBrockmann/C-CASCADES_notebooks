#### Notes on installating packages

Conda vs pip from https://www.continuum.io/content/conda-data-science

Python programmers are probably familiar with pip to download packages from PyPI and manage their requirements. Although, both conda and pip are package managers, they are very different:

    Pip is specific for Python packages and conda is language-agnostic, which means we can use conda to manage packages from any language.
    Pip compiles from source and conda installs binaries, removing the burden of compilation.
    Conda creates language-agnostic environments natively whereas pip relies on virtualenv to manage only Python environments.

Though it is recommended to always use conda packages, conda also includes pip, so you don’t have to choose between the two. For example, to install a python package that does not have a conda package, but is available through pip, just run, for example:

    conda install pip
    pip install aPackage

### Make a map

Cartopy is a Python package for advanced map generation with a simple matplotlib interface.

http://scitools.org.uk/cartopy/index.html

To be installed from

    conda -c scitools cartopy

```python
import matplotlib.pyplot as plt
import cartopy.crs as ccrs

%matplotlib inline
```

```python
plt.figure(figsize=(10, 12))

ax = plt.axes(projection=ccrs.Robinson())
#ax = plt.axes(projection=ccrs.PlateCarree())

#ax.set_extent([-45, 55, 20, 80], ccrs.PlateCarree())
ax.set_global()

ax.stock_img()
ax.coastlines()

linewidth = 3
plt.plot(locationParis.longitude, locationParis.latitude, 'o', transform=ccrs.PlateCarree())
for pos in dfs['locationPos']:
    plt.plot(pos.longitude, pos.latitude, 'ko', transform=ccrs.PlateCarree())
    plt.plot([locationParis.longitude, pos.longitude], [locationParis.latitude, pos.latitude], 
                                     linewidth=linewidth, transform=ccrs.Geodetic())


plt.show()
```

### Make an slippy map

https://github.com/python-visualization/folium
    
to be installed from

    pip install folium

```python
import folium
```

```python
map = folium.Map(location=[0, 0], zoom_start=1, tiles='OpenStreetMap')   # 'OpenStreetMap', 'Stamen Terrain', 'Stamen Toner'

folium.Marker([locationParis.latitude, locationParis.longitude], popup='C-CASCADES Training Workshop - March 2017').add_to(map)

for i in range(0, len(dfs)):
    
    lat = dfs.iloc[i]['locationPos'].latitude
    lon = dfs.iloc[i]['locationPos'].longitude
    name = dfs.iloc[i]['name']

    color = rand_color.generate(luminosity='bright')[0]

    folium.CircleMarker(location=[lat, lon], radius=5,
                    popup=name, color="#666", fill_color=color).add_to(map)
    
    folium.PolyLine([(lat,lon),(locationParis.latitude,locationParis.longitude)], 
                    color=color, opacity=0.9, weight=4).add_to(map)

map
```
