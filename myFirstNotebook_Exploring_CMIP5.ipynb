```python
%%ferret -s 800,800

set mode linecolors:20

let a =  FGCO2[i=@ave,j=@ave]*1E9

let/title="1pctCO2" a1 = a[m=1]
let/title="esmFixClim1" a2 = a[m=2]
let/title="esmFixClim1" a3 = a[m=3]

! keep color black as 1
ppl color 2, 0, 50, 100, 50
ppl color 3, 100, 50, 0, 50
ppl color 4, 80, 0, 80, 50
ppl color 5, 100, 0, 0, 50
ppl color 20, 70, 70, 70

plot/color/grat=(dash,color=20)/key=title/thick=3 a1, a2, a3

! plot the members average
plot/over/color=5/thick=3/dash=(0.1,0.1,0.1,0.1)/title="Average" a[m=@ave]
```
